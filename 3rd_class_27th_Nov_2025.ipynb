```python
def eye_exam_covered(age, qualifying_condition, time_since_last_exam):
    # assumption
    # age as int
    # qualifyinf_condition as a boolean , 
    # True if you have eligible medical condition, False otherwise
    # time_since_last_exam int, number of months since last examination
    if 0 <= age <= 19:
        if time_since_last_exam >= 12:
            return "You are eligible for 1 major eye exam and any minor assessments needed"
        else:
            return "It has't been 12 months since your last major eye exam but you can have a minor assessments"
    elif 20 <= age <=64:
        if qualifying_condition:
            if time_since_last_exam >= 12:
                return "You are eligible for 1 major exam and 2 additional follow up minor assessments"
            else:
                return("It hasn't been 12 months since your last major eye exam, but you have follow up for minor assessments")
        else:
            return "You don't have eligible medical conditions"
    elif age >= 65:
        if time_since_last_exam >=18:
            return "You are eligible for 1 major eye exam and 2 minor foolowup assessments"
        else:
            return "It hasn't been 18 months since last major eye exam, but you have 2 follow up minor assessments"
    else:
        return "Invalid age input" 
```
