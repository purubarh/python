```python
def calc_sales_tax(price, tax_rate=0.13):
    sales_tax = price * tax_rate
    return sales_tax
```

```python
calc_sales_tax(500)
```

```python
calc_sales_tax(500, 0.08)
```

```python
def calc_total_bill(price, tax_rate=0.13, tip_rate=0.15):
    # Docstring Used in below line
    '''Calculate total bill with tax and tip.'''
    tax = price * tax_rate
    tip = price * tip_rate
    total = price + tax + tip
    return tax, tip, total
```

```python
calc_total_bill(200)
```

```python
calc_total_bill(200, 0.18)
```

```python
calc_total_bill(200, tip_rate=0.20)
```

```python
calc_total_bill(200,tip_rate=0.10)
```
